```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = []

with open("./data/arjun_drqa.txt") as f:
    for line in f:
        if line.startswith("Epoch"):
            data.append(line)
```

```python
data
```

```python
loss_arr = [float(d.split(":")[1].split("|")[0].strip().split()[1]) for d in data if "loss" in d]
```

```python
lr_arr = [float(d.split(":")[1].split("|")[1].strip().split()[1]) for d in data if "loss" in d]
```

```python
acc_arr = [float(d.split(":")[1].split("|")[1].strip().split()[1]) for d in data if "end_acc" in d]
```

```python
f1_arr = [float(d.split(":")[1].split("|")[3].strip().split()[1]) for d in data if "f1" in d]
```

```python
sns.set_style("darkgrid")
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
ax1 = sns.lineplot(x=[i for i in range(1, 47)], y=loss_arr)#, label = "Loss")
#ax2 = sns.lineplot(x=[i for i in range(1, 47)], y=lr_arr, label = "Learning Rate")
ax1.set(xlabel = "Epochs", ylabel = "Validation loss", title = "Validation loss over epochs")
plt.show()
```

```python
ax1 = sns.lineplot(x=[i for i in range(1, 47)], y=acc_arr)#, label = "Loss")
#ax2 = sns.lineplot(x=[i for i in range(1, 47)], y=lr_arr, label = "Learning Rate")
ax1.set(xlabel = "Epochs", ylabel = "Validation accuracy", title = "Validation accuracy over epochs")
plt.show()
```

```python
ax1 = sns.lineplot(x=[i for i in range(1, 47)], y=f1_arr)#, label = "Loss")
#ax2 = sns.lineplot(x=[i for i in range(1, 47)], y=lr_arr, label = "Learning Rate")
ax1.set(xlabel = "Epochs", ylabel = "F1 score", title = "F1 score over epochs")
plt.show()
```

```python
ax1 = sns.lineplot(x=[i for i in range(1, 47)], y=lr_arr)#, label = "Loss")
#ax2 = sns.lineplot(x=[i for i in range(1, 47)], y=lr_arr, label = "Learning Rate")
ax1.set(xlabel = "Epochs", ylabel = "Learning rate", title = "Learning rate over epochs")
plt.show()
```
